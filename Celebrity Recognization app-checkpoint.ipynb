```python
def get_copied_img_if_2_eye(img_path):
    img=cv2.imread(img_path)
    if img is None:
        exit(1)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        gray_i=gray[y:y+h,x:x+w]
        color_i=img[y:y+h,x:x+w]
        eyes=eye_cascade.detectMultiScale(gray_i)
        if len(eyes)>=2:
            return color_i
    
```

```python
import os
```

```python
img_dirs=[]
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
if os.path.exists(path_of_crop_data):
    shutil.rmtree(path_of_crop_data)
os.mkdir(path_of_crop_data)    
```

```python
cropped_image_dirs=[]
celebrity_file_names_dict={}
for img_dir in img_dirs:
    count=1
    celebrity_name=img_dir.split('/')[-1]
    print(celebrity_name)
    celebrity_file_names_dict[celebrity_name]=[]
    for entry in os.scandir(img_dir):
        color_i=get_copied_img_if_2_eye(entry.path)
        if color_i is not None:
            cropped_folder=path_of_crop_data + celebrity_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
            cropped_file_name=celebrity_name + str(count) + ".png"
            cropped_file_path=cropped_folder + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path,color_i)
            celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
            count+=1
                
    
```

```python
import pywt
```

```python
def w2d(img,mode='haar',level=1):
    image_arr=img
    #convert to grayscale
    image_arr=cv2.cvtColor(image_arr,cv2.COLOR_RGB2GRAY)
    #convert to float
    image_arr=np.float32(image_arr)
    image_arr/=255
    #Compute the coefficients with the help of pwt.wavedec2
    #Takes input as gray scale image,mode,level
    #returns the tupple of coefficient
    coeff=pywt.wavedec2(image_arr,mode,level=level)
    #process coefficient
    coeffs_H=list(coeff)
    coeffs_H[0]*=0
    #The modified coefficients (coeffs_H) are used to perform the inverse wavelet transform using pywt.waverec2
    #Reconstruction of image
    image_arrH=pywt.waverec2(coeffs_H,mode)
    image_arrH*=255
    image_arrH=np.uint8(image_arrH)
    return image_arrH
    
    
```

```python
d={}
count=0
for cel_name in celebrity_file_names_dict.keys():
    d[cel_name]=count
    count+=1
d    
    
```

```python
X=[]
y=[]
for cele_name,val in celebrity_file_names_dict.items():
    for v in val:
        img=cv2.imread(v)
        if img is None:
            continue
        scalled_img=cv2.resize(img,(32,32))
        img_w2d=w2d(img,'db1',5)
        scalled_img_w2d=cv2.resize(img_w2d,(32,32))
        combined_img=np.vstack((scalled_img.reshape(32*32*3,1),scalled_img_w2d.reshape(32*32,1)))
        X.append(combined_img)
        y.append(d[cel_name])
```
